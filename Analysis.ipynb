# Mini Project

```python
#Data Loading
import pandas as pd
import numpy as np
moviedata = pd.read_csv('tmdb-movies.csv')
moviedata.info()
moviedata.head()
```

```python
# Data Cleaning
moviedata.drop_duplicates(inplace=True)
moviedata.info()
```

```python
moviedata.dropna(subset=['genres'], inplace=True)  
moviedata.info()
```

```python
nmd = moviedata[['popularity','budget','revenue', 'original_title','runtime','genres','release_year','cast','production_companies']]
nmd.head()
```

```python
 nmd= nmd[nmd['budget'] != 0]
 nmd = nmd[nmd['revenue'] != 0]
 nmd = nmd[nmd['runtime'] != 0]
 nmd
```

# 4-Which are the movies with most and least earned revenue

```python
#highest
nmd.sort_values(by = 'revenue',ascending = False).loc[:,['revenue', 'original_title']][0:10]
```

```python
#least
nmd.sort_values(by = 'revenue',ascending = True).loc[:,['revenue', 'original_title']][0:10]
```

# 6-Name any 3 production companies which have invested money in worse revenue movies

```python
nmd.sort_values(by = 'revenue',ascending = True).loc[:,['revenue', 'original_title','production_companies']][0:10]
```

# 1-Which are the movies with the third lowest and third highest budget?

```python
nmd.sort_values(by = 'budget',ascending = False).loc[:,['budget', 'original_title']][0:3]
```

```python
nmd.sort_values(by = 'budget',ascending = True).loc[:,['budget', 'original_title']][0:3]
```

# 5-What is the average runtime of movies in the year 2006?

```python

#print(f"average runtime of movies of 2006 in : {nmd['runtime'].mean():4f}mins")
#nmd['release_year'].value_counts()[0:]
nmd_2006 = nmd[nmd['release_year'] == 2006]
total_runtime = 0
for i in range(nmd_2006.shape[0]):
    total_runtime += int(nmd_2006['runtime'].values[i])
    
avg_time = total_runtime/nmd_2006.shape[0]
print("Avarage runtime of movoes in the year 2006 is :",avg_time)
```
